```python
from pathlib import Path
import pandas as pd

data_root = Path("data")
rows = []

for ds_dir in sorted(p for p in data_root.iterdir() if p.is_dir()):
    answer_path = ds_dir / "answer.csv"
    truth_path = ds_dir / "truth.csv"
    if not answer_path.exists() or not truth_path.exists():
        continue

    answer = pd.read_csv(answer_path, header=None, encoding="utf-8-sig")
    pd.read_csv(truth_path, header=None, encoding="utf-8-sig")  # ensure BOM handled

    rows.append({
        "dataset": ds_dir.name,
        "objects": answer.iloc[:, 0].nunique(),
        "sources": answer.iloc[:, 1].nunique(),
        "claims": len(answer),
    })

summary_df = pd.DataFrame(rows).sort_values("dataset").reset_index(drop=True)
summary_df
```
